# Notebook for scraping NHL data using the `chicken_nhl` scrape functions

## Overview

We'll be scraping data from the 2016-17 playoffs for our tutorials.

1. First we'll scrape the schedule to get the game IDs
2. Then we'll scrape the play-by-play data using the given game IDs
3. Finally, we'll save the CSV file to the processed folder

The purpose of this exercise to demonstrate an archetypal process for using `chicken_nhl`
Using this notebook to scrape the data is optional. Feel free to skip this step use the processed csv file

## Housekeeping

### Dependencies

```python
import pandas as pd
import numpy as np

from pathlib import Path

import os

from chickenstats.chicken_nhl import scrape_schedule, scrape_pbp
```

## Getting game IDs via the schedule

We'll start by scraping the schedule for the 2016-17 playoffs using the `scrape_schedule()` function and keyword arguments

```python
schedule = scrape_schedule(2016, game_types = ['P'])
```

Then we'll get the game IDs from the game ID column

```python
game_ids = schedule.game_id
```

## Scraping PBP

Now, we'll just feed the game IDs to the `scrape_pbp()` function. It will take about five minutes, feel free to go get some coffee :)

```python
df = scrape_pbp(game_ids)
```

## Saving to CSV

```python
folder_stem = Path('./processed/chicken_nhl/')

if os.path.exists(folder_stem) == False:
    
    os.makedirs(folder_stem)

file_path = folder_stem / 'pbp_chicken.csv'

df.to_csv(file_path)
```
